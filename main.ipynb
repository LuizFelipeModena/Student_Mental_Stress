```python
import pandas as pd
```

```python
df = pd.read_csv('Student_Mental_Stress_and_Coping_Mechanisms.csv')

df.head()
```

```python
df.info()
```

Descrição das colunas:

- Student ID: Código ID dos alunos.

- Age: Idade do estudante (entre 18 e 30 anos).
 
- Gender: Male, Female, Non-binary, Prefer not to say
 
- Academic Performance: Média de notas do aluno (De 0.0 a 4.0).
 
- Study Hours Per Week: Tempo de estudos por semana (De 0 a 60 horas).
 
- Social Media Usage: Média de horas por dia que o aluno usa rede social.
 
- Sleep Duration: Tempo de sono do aluno por noite em horas.
 
- Physical Exercise: Tempo médio de exercício físico que o estudante pratica por semana em horas.
 
- Family Support: Autorrelato do nível de suporte da família (1 = pouco suporte, 5 = muito suporte).
 
- Financial Stress: Autorrelato do estresse financeiro (1 = sem estresse, 5 = muito estresse).
 
- Peer Pressure: Autorrelato das pressão causada por colegas (1 = pouca pressão, 5 = muita pressão).
 
- Relationship Stress: Autorrelato do nível de estresse em relacionamento (1 = pouco estresse, 5 = muito estresse).
 
- Mental Stress Level: Autorrelato do nível de estresse mental (1 = sem estresse, 10 = estresse extremo).
 
- Counseling Attendance: Se o aluno passou por aconselhamento (Yes/No).
 
- Diet Quality: Autorrelato da qualidade da dieta (1 = dieta ruim, 5 = dieta excelente).
 
- Stress Coping Mechanisms: Estretagéia que o aluno usa para alívio de estresse, incluindo: Talking to Family, Reading, Yoga, Exercise, Meditation, And many more…
 
- Cognitive Distortions: Autorrelato do nível de distorção cognitiva (1 = sem distorções, 5 = distorções severas).
 
- Family Mental Health History: Se o estudante te histórico de problemas mentais na família.
 
- Medical Condition: Se o aluno tem alguma condição mental pré-existente que possa afetar sua saúde mental.
 
- Substance Use: Autorrelato do nível de uso de substâncias (1 = não faz uso, 5 = faz uso pesado).
